```julia
using DelimitedFiles
using PyPlot
using Random
include("RASJulia.jl")

const X = readdlm("../Data/ToyData.csv", ',',Float64, '\n');
const N = size(X,1);

# Parameters
const bw = 1.;
const epsilon = 10^(-10);
const updating = true;
const lambda  = 10^(-6);
const NbFourierF = 4000;
const c = 200*epsilon;
const t =  0.;
```

```julia
idS = zeros(Int64,N,1);
nbSamples = 0;
idS, nbSamples = @time RASJulia(X, bw , lambda, c , t, epsilon, updating, NbFourierF);
Sample = idS[1:nbSamples];

scatter(X[:,1],X[:,2],alpha=0.5)
scatter(X[Sample,1],X[Sample,2])
```
